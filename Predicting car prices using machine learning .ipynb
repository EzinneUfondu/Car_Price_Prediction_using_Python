# Data Preprocessing & Feature Engineering

```python
# since we have already replaced "?" with "nan", lets go ahead to identify missing columns.
```

```python
df.isnull().sum()
```

```python
df["price"][5]
```

```python
df["normalized-losses"].dtype
```

```python
df["normalized-losses"].shape
```

```python
# Replacing with mean for numeric columns

numeric_cols = ["normalized-losses", "bore", "stroke", "horsepower", "peak-rpm"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
# Replacing with mode for categorical column

df["num-of-doors"] = df["num-of-doors"].astype(str)
df["num-of-doors"].fillna(df["num-of-doors"].mode()[0], inplace=True)
```

# Convert Data Types`

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
```

```python
# Remove rows where price is missing

df = df[df["price"].notna()]
```

```python
print(df.dtypes)
```

# Feature Engineering

```python
df["city-mpg"] = pd.to_numeric(df["city-mpg"], errors="coerce")
df["city-L/100km"] = 235 / df["city-mpg"]
```

# Model Development (Regression)

```python
# Import Your Libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Choose features

x = df[["engine-size", "horsepower", "curb-weight", "width"]]
y = df["price"]
```

```python
# Spliting the Data into 80% Train / 20% Test
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

# Predictions

```python
y_pred_train = lr.predict(x_train)
y_pred_test = lr.predict(x_test)
```

```python
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

mse_train, mse_test
```

```python
R2_Score = r2_score(y_test, y_pred_test)
R2_Score
```

# Random Forest Regressor
#Train the Random Forest Mode

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf.fit(x_train, y_train)
```

```python
# Predict on the Test Set
```

```python
rf_pred = rf.predict(x_test)
```

```python
# Evaluate Random Forest
```

```python
from sklearn.metrics import r2_score, mean_squared_error

rf_r2 = r2_score(y_test, rf_pred)

rf_r2
```

```python
rf_mse = mean_squared_error(y_test, rf_pred)
rf_mse
```

# Classification Model
# Convert price into categories: Low Price, Medium Price, High Price

```python
df['price-category'] = pd.qcut(
    df['price'], 
    3, 
    labels=['Low', 'Medium', 'High']
)
```

```python
# Lets Prepare x and y for Classification
```

```python
x_class = df[["engine-size", "horsepower", "curb-weight", "width"]]
y_class = df["price-category"]
```

# Train/Test Split with 70/30

```python
x_train_c, x_test_c, y_train_c, y_test_c = train_test_split(
    x_class, y_class, test_size=0.3, random_state=42
)
```

# Train Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression

clft = LogisticRegression(max_iter=500)
clft.fit(x_train_c, y_train_c)
```

# Prediction

```python
y_pred_c = clft.predict(x_test_c)
```

# Lets Evaluate using Accuracy and the Confusion Matrix

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accu = accuracy_score(y_test_c, y_pred_c)
comt = confusion_matrix(y_test_c, y_pred_c)

print(accu)

print(comt)
```
